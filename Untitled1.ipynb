```python
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
name = "custom prior less steps" # To change to match data

data_dir = "./run_outputs/regression_rollout/"
data_file = name + ".result_stream.pickle"

#### Creating the Dataset

with open(data_dir + 'ys' + data_file, 'rb') as f:
    ys = pickle.load(f)
with open(data_dir + 'ds' + data_file, 'rb') as f:
    ds = pickle.load(f)
n_data, seq_len, n = ys.shape
output_dim = ds.shape[-1]
input_dim = n
p = output_dim//n
```

```python
with open(data_dir + 'plot_ys' , 'rb') as f:
    plot_ys = pickle.load(f)
with open(data_dir + 'plot_ds', 'rb') as f:
    plot_ds = pickle.load(f)
with open(data_dir + 'plot_out', 'rb') as f:
    plot_out = pickle.load(f)    
```

```python
plt.plot(plot_ds[:100,0,0])
```

```python
plt.plot(ds[:100,0,0])
```

```python
plot_dir = "./run_outputs/Data_visualization_ regression_rollout/"
for i in range(output_dim):
    x = ys[:,0,0].numpy()
    y = ys[:,0,1].numpy()
    z = ds[:,1,i].numpy()
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    ax.plot_trisurf(x, y, z, cmap=plt.cm.Spectral)

    plt.savefig(data_dir + 'data_ds[{}] against ys.png'.format(i))
```

```python
for i in range(output_dim):
    x = plot_ys[:,0,0].detach().numpy()
    y = plot_ys[:,0,1].detach().numpy()
    z = plot_out[:,1,i].detach().numpy()
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    ax.plot_trisurf(x, y, z, cmap=plt.cm.Spectral)

    plt.savefig('out[{}] against ys.png'.format(i))
```

```python
plot_dir = "./run_outputs/Data_visualization_ regression_rollout/"
fig = plt.figure(figsize=(11, 8))
for i in range(output_dim):
    x = plot_ds[:,1,i].detach().numpy()
    y = plot_out[:,1,i].detach().numpy()
    plt.subplot(4,3,i+1)
    plt.scatter(x,y)
plt.savefig(plot_dir + 'output against ds second_seq.png')
```
